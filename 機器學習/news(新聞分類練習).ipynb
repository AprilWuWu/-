```
import zipfile
with zipfile.ZipFile('drive/My Drive/AI/機器學習/datasets/chinese_news_trans.zip', 'r') as zip_ref:
    zip_ref.extractall('')

with zipfile.ZipFile('drive/My Drive/AI/機器學習/datasets/chinese_news_test.zip', 'r') as zip_ref:
    zip_ref.extractall('')
```

```
import glob
import os
import pandas as pd


def get_df(target_dn):
  data = {
      'path': [],
      'label': []
  }

  base_path = ''
  p = os.path.join(base_path, target_dn, '*')
  for dn in glob.glob(p):
    name = os.path.split(dn)[-1]
    fn = os.path.join(dn,'*.[tT][xX][tT]')
    fns = glob.glob(fn)

    data['label'] = data['label'] + [name] * len(fns)

    data['path'] = data['path'] + fns


  return pd.DataFrame(data)
```

```
train_df = get_df('chinese_news_trans')
test_df = get_df('chinese_news_test')
```

```
def read_content(fn):
  with open(fn, 'r', encoding='utf-8') as f:
    content = f.read()
  return content

train_df['content'] = train_df['path'].apply(read_content)
test_df['content'] = test_df['path'].apply(read_content)
```

```
# Series.replace(轉換字典)
# {'李白':0, '杜甫':1}

# unique/ value_counts
u = train_df['label'].unique()
trans = { n:i for i, n in enumerate(u) }
trans_r = { i:n for i, n in enumerate(u) }
y_train = train_df['label'].replace(trans)
y_test = test_df['label'].replace(trans)
```

```
import jieba
#  Series.apply(func_name)
from urllib.request import urlretrieve
url = 'https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big'
urlretrieve(url, 'dict.txt.big')
jieba.set_dictionary('dict.txt.big')

def news_cut(p):
  return ' '.join(jieba.cut(p))

x_train = train_df['content'].apply(news_cut)
x_test = test_df['content'].apply(news_cut)
x_test
```

```
# countvectorizer
# fit  找出應該有幾個欄位
# transform 按照fit找出的欄位數數

# 訓練資料 = fit = transform
# 測試資料 = transform

from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_count = vec.fit_transform(x_train)
x_test_count = vec.transform(x_test)
```

```
# 2731(資料筆數) * 52294(欄位)
# sparse matrix 很多位置都是0，只存非0的值
x_train_count

# print(x_train_count)
# (0, 16053) 1  第一首詩在第16053個欄位出現一次

# 欄位對應
vec.vocabulary_

# 檢查特殊符號是否在裡面
# vec.vocabulary_['。']
```

# stopwords 自己停用某些欄位  Ex:我  這個欄位，出現很多次，可是大家都用的次數都差不多，所以不需要自己刪除

$$ P(\frac{我}{杜甫}) \approx P(\frac{我}{李白})$$  

```
from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()
clf.fit(x_train_count, y_train)
```

## https://scikit-learn.org/stable/modules/naive_bayes.html
Complement Navie Bayes (用在某個類別數量特別少的時候)

Gaussian Navie Bayes (用高斯曲線算出每個點的分布機率)

Bernoulli Navie Bayes (應付二值型的資料)


```
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_count)
accuracy_score(pre, y_test)
```

```
# confustion_matrix (混淆矩陣)
# https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)

pd.DataFrame(mat, 
             columns=[n+'(預測)' for n in u],
             index=[n+'(真實)' for n in u])
```

```
test = input('輸入一則新聞:')
# 江上日多雨
count = vec.transform([news_cut(test)]) # 剛剛進去是一群，所以現在也要給陣列
# x_train

answer = clf.predict(count)[0]
proba = clf.predict_proba(count)[0] # 機率
print(proba)
for name, p in zip(u, proba):
  print(name, '的機率', p)
print('答案:', trans_r[answer])
```
