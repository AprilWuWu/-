```
import numpy as np
samples = np.random.normal(loc=10, scale=2, size=10000)
```

---
# 常態分布

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.distplot(samples)
```

```
s = samples[samples > 0] + 1 
sns.distplot(s)
```

---
# 右偏

```
from scipy.stats import skew
new = s ** 5
sns.distplot(new)
print('skew:', skew(new))
```

```
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.boxcox_normmax.html
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.special.boxcox.html

from scipy.stats import boxcox_normmax, boxcox
lamda = boxcox_normmax(new) # 偏度
print(lamda)
new2 = boxcox(new, lamda)
sns.distplot(new2)
print(skew(new2))
```

---
# 左偏


```
new = s ** (1/5)
sns.distplot(new)
print('skew:', skew(new))
```

```
from scipy.stats import boxcox_normmax, boxcox
lamda = boxcox_normmax(new) # 偏度
print(lamda)
new2 = boxcox(new, lamda)
sns.distplot(new2)
print(skew(new2))
```

---
# 極度右偏

```
# https://numpy.org/doc/stable/reference/generated/numpy.power.html
new = np.power(2, s)
sns.distplot(new)
print('skew:', skew(new))
```

```
import numpy as np
from scipy.stats import boxcox_normmax, boxcox
lamda = boxcox_normmax(new) # 偏度
print(lamda) # 趨近於0，可以自己做log
# new2 = boxcox(new, lamda)
new2 = np.log(new)
sns.distplot(new2)
print(skew(new2))
```
